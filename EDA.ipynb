```python
import pandas as pd

# Read the Excel file
data = pd.read_excel('dataset.xlsx')

# Display the first few rows
print(data.head())
```

```python
import matplotlib.pyplot as plt

# Visualizing the distribution of the target variable
plt.figure(figsize=(6, 4))
data['target'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title('Target Variable Distribution')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()

# Summary statistics of numeric features
summary_stats = data.describe()

# Correlation heatmap for feature relationships
correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', aspect='auto')
plt.colorbar(label='Correlation Coefficient')
plt.title('Feature Correlation Heatmap')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.tight_layout()
plt.show()

summary_stats
```

```python
# Feature distribution plots
import seaborn as sns
import math

# Number of features to plot
num_features = len(data.select_dtypes(include=[float, int]).columns)

# Calculate grid dimensions
ncols = 5  # Set the number of columns per row
nrows = math.ceil(num_features / ncols)

plt.figure(figsize=(20, nrows * 4))  # Dynamically adjust figure height
for i, column in enumerate(data.select_dtypes(include=[float, int]).columns, 1):
    plt.subplot(nrows, ncols, i)
    sns.histplot(data[column], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

```

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Features to plot
features = data.select_dtypes(include=[float, int]).columns.drop('target')
num_features = len(features)

# Calculate grid dimensions
ncols = 5  # Fixed number of columns
nrows = math.ceil(num_features / ncols)  # Calculate required rows

# Adjust figure size dynamically
plt.figure(figsize=(ncols * 4, nrows * 4))  # Each subplot gets 4x4 size

# Create boxplots
for i, column in enumerate(features, 1):
    plt.subplot(nrows, ncols, i)
    sns.boxplot(x='target', y=column, data=data, palette='Set2')
    plt.title(f'{column} vs Target')

# Layout adjustments
plt.tight_layout()
plt.show()
```
