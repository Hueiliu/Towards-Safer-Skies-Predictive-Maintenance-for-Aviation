```python
import dask.dataframe as dd
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

```python
df = df.drop(columns=['timestep', 'cluster'])
```

```python
df = df.fillna(df.mean())
df.columns
```

```python
df['Voltage_Difference'] = df['volt1'] - df['volt2']
df['Battery_Current_Ratio'] = df['amp1'] / df['amp2']
df['Total_Fuel_Quantity'] = df['FQtyL'] + df['FQtyR']
df['Fuel_Imbalance'] = abs(df['FQtyL'] - df['FQtyR'])
df['Average_Cylinder_Head_Temp'] = (df['E1 CHT1'] + df['E1 CHT2'] + df['E1 CHT3'] + df['E1 CHT4']) / 4
df['Exhaust_Gas_Temp_Spread'] = df[['E1 EGT1', 'E1 EGT2', 'E1 EGT3', 'E1 EGT4']].max(axis=1) - df[['E1 EGT1', 'E1 EGT2', 'E1 EGT3', 'E1 EGT4']].min(axis=1)
df['Oil_Pressure_to_Temp_Ratio'] = df['E1 OilP'] / df['E1 OilT']
df['Normalized_Air_Speed'] = df['IAS'] / df['AltMSL']
df['Engine_Efficiency'] = df['E1 FFlow'] / df['E1 RPM']
df['Thermal_Efficiency'] = df[['E1 EGT1', 'E1 EGT2', 'E1 EGT3', 'E1 EGT4']].sum(axis=1) / df[['E1 CHT1', 'E1 CHT2', 'E1 CHT3', 'E1 CHT4']].sum(axis=1)
df['Temp_Stress'] = df['OAT'] * df[['E1 CHT1', 'E1 CHT2', 'E1 CHT3', 'E1 CHT4']].mean(axis=1)
df['Temp_Imbalance'] = df[['E1 CHT1', 'E1 CHT2', 'E1 CHT3', 'E1 CHT4']].max(axis=1) - df[['E1 CHT1', 'E1 CHT2', 'E1 CHT3', 'E1 CHT4']].min(axis=1)
df.columns
```

```python
import numpy as np

def replace_inf_with_min_max(df):
    for col in df.columns:
        if np.issubdtype(df[col].dtype, np.number):  # Process only numeric columns
            # Replace -inf with column minimum (ignoring inf values)
            if np.isinf(df[col]).any():
                col_min = df.loc[~np.isinf(df[col]), col].min()
                df[col] = df[col].replace(-np.inf, col_min)

                # Replace +inf with column maximum (ignoring inf values)
                col_max = df.loc[~np.isinf(df[col]), col].max()
                df[col] = df[col].replace(np.inf, col_max)
    return df

# Apply the function to your DataFrame
df = replace_inf_with_min_max(df)

# Display updated DataFrame
print(df)
```

```python
df.to_csv('dataset.csv', index=False)
```
